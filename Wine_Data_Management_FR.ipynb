# Fonction Et Test




```python
#Fonction de test d'unicité clé primaire


def unicited(data,col) :
        if len(data) == data[col].nunique():
            print("C'est bien une clé primaire")
        else :
            print("Ce n'est pas une clé primaire")
```

```python
#Choix et test de l'unicité de la clé primaire de la table erp

unicited(erp,'product_id')
```

```python
#Choix et test de l'unicité de la clé primaire de la table liaison

unicited(liaison,'product_id')
```

# Solution mise en place pour les outliers



```python
#Chasse aux outliers


import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot
vata.boxplot(column='price', vert=True)
plt.show()
```

```python
#Méthode interquartile pour outliers

#Calcul des 3 quartiles

rocsol = vata['price'].quantile([0.25, 0.5, 0.75])
rocsol
```

```python
#Méthode pour obtenir les outliers

#Calcul de Q1 et Q3

firstq = vata['price'].quantile(0.25)
thirdq = vata['price'].quantile(0.75)

#Ecart-interquartile
quartz = thirdq - firstq
quartz

#Référence pour outliers
matz = quartz*1.5
matz

infz = firstq - matz
supz = thirdq + matz

print(quartz,matz,firstq,thirdq)
print("Les valeurs inférieurs à",infz,"et supérieurs à",supz,"sont des outliers")
```

```python
#Mise en évidence des outliers

prouvo = vata[vata['price'] <= supz]
drouvo = vata[vata['price'] >= supz]
```

```python
#Valeurs inférieurs ou égales à la moustache

prouvo.info()
prouvo.head()
```

```python
#Valeurs suppérieures ou égales à la moustache

drouvo.info()
drouvo.head()
```

```python
#Représentation graphiques avec mise en évidence des outliers

# Scatterplot

plt.scatter(prouvo['product_id'],prouvo['price'],color='green', s=40, label='normal')
plt.scatter(drouvo['product_id'],drouvo['price'],color='red', s=40, label='outliers')
plt.title("Result")
plt.legend(loc=2)
plt.xlabel("Product_id")
plt.ylabel("Price")
plt.show()
```
